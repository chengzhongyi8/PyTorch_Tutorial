## PyTorch卷积模块
### 卷积层
卷积在 PyTorch 中有两种方式，一种是 `torch.nn.Conv2d()`，一种是 `torch.nn.functional.conv2d()`，这两种形式本质都是使用一个卷积操作。

这两种形式的卷积对于输入的要求都是一样的，首先需要输入是一个 `torch.autograd.Variable()` 的类型，大小是 (batch, channel, H, W)，其中 batch 表示输入的一批数据的数目，第二个是输入的通道数，一般一张彩色的图片是 3，灰度图是 1，而卷积网络过程中的通道数（feature map的个数）比较大，会出现几十到几百的通道数，H 和 W 表示输入图片的高度和宽度，比如一个 batch 是 32 张图片，每张图片是 3 通道，高和宽分别是 50 和 100，那么输入的大小就是 (32, 3, 50, 100)。

**卷积之后的图像大小为 (W-F+2\*P)/S + 1，并且向下取整， F是kernel size，P是padding的大小，S是strides。**

以下举例说明。

```python
import numpy as np
import torch 
from torch import nn
from torch.autograd import Variable
import torch.nn.functional as F
from PIL import Image
import matplotlib.pyplot as plt
from torchsummary import summary
%matplotlib inline
```

```python
im = Image.open('./Images/cat.png').convert('L') # 读入一张灰度图的图片，open()读入的是RGB的图像格式，用convert将其转换成L，也即灰度格式
im = np.array(im, dtype='float32') # 将其转换为一个矩阵
print(im.shape)

# 可视化图片
plt.imshow(im.astype('uint8'), cmap='gray')
```

```python
Image.open('./Images/cat.png').mode
```

```python
# 将图片矩阵转化为 pytorch tensor，并适配卷积输入的要求
im = torch.from_numpy(im.reshape((1, 1, im.shape[0], im.shape[1])))  ## 一个样本，单通道数据
```

### 利用卷积计算图像梯度
* 图像梯度的计算用于图像边缘检测，就是检测出变化比较大的地方。边缘检测对于图像中的物体分割和识别很重要，因为它可以提取出比较重要的特征。
* Sobel filter以离散的方式（有限差分）计算图像梯度，H算子用于计算x方向的梯度，V算子用于计算y方向的梯度。

<img src="./Images/Sobel_filter1.jpg", width=350, heigth=300>
* 关于图像梯度计算以及各种算子的介绍可参考 http://www.sci.utah.edu/~gerig/CS6640-F2014/Materials/EdgeDetection-Bebis-UNR-I.pdf

```python
conv1 = nn.Conv2d(1,1,3,bias=False)  ## 卷积核的weight会被随机初始化,但这个weight不能满足特定的需求——检测边缘。默认stride=1,padding=0

Gx = np.array([[-1,0,1],[-2,0,2],[-1,0,1]],dtype='float32')
Gy = np.array([[1,2,1],[0,0,0],[-1,-2,-1]],dtype='float32')
Gx = Gx.reshape((1, 1, 3, 3)) # 适配卷积的输入输出
Gy = Gy.reshape((1, 1, 3, 3))
conv1.weight.data = torch.from_numpy(Gx) # 给卷积的 kernel 赋值

print('Shape before conv:',im.shape)
edge1 = conv1(Variable(im)) # 作用在图片上
print('Shape after conv:',edge1.shape)

edge1 = edge1.data.squeeze().numpy() # 将输出转换为图片的格式


fig,ax = plt.subplots(1,2,figsize=(10,4))
ax[0].imshow(edge1, cmap='gray') ## 可视化边缘检测之后的结果

conv1.weight.data = torch.from_numpy(Gy)  # 再用Gy计算y向的梯度 
edge2 = conv1(Variable(im)) # 作用在图片上
edge2 = edge2.data.squeeze().numpy()
ax[1].imshow(edge2, cmap='gray')
```

```python
## 用F.conv2d实现卷积运算也是一样的效果
weight_x = Variable(torch.from_numpy(Gx))

edge3 = F.conv2d(Variable(im), weight_x) # 作用在图片上,F.conv2d(image, kernel)
edge3 = edge3.data.squeeze().numpy() # 将输出转换为图片的格式

fig,ax = plt.subplots(1,2,figsize=(10,4))
ax[0].imshow(edge3, cmap='gray')

weight_y = Variable(torch.from_numpy(Gy))
edge4 = F.conv2d(Variable(im), weight_y) # 作用在图片上,F.conv2d(image, kernel)
edge4 = edge4.data.squeeze().numpy()

ax[1].imshow(edge4, cmap='gray')
```

可以看到两种形式能够得到相同的效果，不同的地方在于，使用 `nn.Conv2d()` 相当于直接定义了一层卷积网络结构，而使用 `torch.nn.functional.conv2d()` 相当于定义了一个卷积的操作，所以使用后者需要再额外去定义一个 weight，而且这个 weight 也必须是一个 Variable，而使用 `nn.Conv2d()` 则会帮我们默认定义一个随机初始化的 weight，如果我们需要修改，那么取出其中的值对其修改，如果不想修改，那么可以直接使用这个默认初始化的值，非常方便。

**实际使用中我们基本都使用 `nn.Conv2d()` 这种形式**

### 池化层
卷积网络中另外一个非常重要的结构就是池化，这是利用了图片的下采样不变性，即一张图片变小了还是能够看出了这张图片的内容，而使用池化层能够将图片大小降低，非常好地提高了计算效率，同时池化层也没有参数。池化的方式有很多种，比如最大值池化，均值池化等等，在卷积网络中一般使用最大值池化。

在 PyTorch 中最大值池化的方式也有两种，一种是 `nn.MaxPool2d()`，一种是 `torch.nn.functional.max_pool2d()`，他们对于图片的输入要求跟卷积对于图片的输入要求是一样的。

**池化之后得到的图片大小为 (W-F)/S + 1，并且向下取整。**

```python
# 使用 nn.MaxPool2d
pool1 = nn.MaxPool2d(2, 2) # kernel size=2, stride=2
print(f'before max pool, image shape: {im.shape[2]} x {im.shape[3]}')
small_im1 = pool1(Variable(im))
small_im1 = small_im1.data.squeeze().numpy()
print(f'after max pool, image shape: {small_im1.shape[0]} x {small_im1.shape[1]}')
```

```python
plt.imshow(small_im1, cmap='gray')
```

可以发现池化只是使图片的大小变小了，但是并没有改变图像的内容。

```python
# F.max_pool2d
print(f'before max pool, image shape: {im.shape[2]} x {im.shape[3]}')
small_im2 = F.max_pool2d(Variable(im), 2, 2)
small_im2 = small_im2.data.squeeze().numpy()
print('after max pool, image shape: {small_im1.shape[0]} x {small_im1.shape[1]}')
plt.imshow(small_im2, cmap='gray')
```

**跟卷积层一样，实际使用中，我们一般使用 `nn.MaxPool2d()`**

## 简单的CNN案例
假设我们有图像数据，size为[100,3,64,64]

一方面要自己会算每一层的feature map的大小，另一方面也要在搭好网络之后就验证一下输出的size是否正确。

```python
class SimpleCNN(nn.Module):  ## 使用nn.Module搭建网络，并且内嵌了nn.Sequential来构建block
    def __init__(self):
        super(SimpleCNN,self).__init__()
        layer1 = nn.Sequential()  ## 通过add_module可以比较方便的各层命名
        layer1.add_module('conv1',nn.Conv2d(3, 32, 3, 1,padding=1))  #[100, 32, 64, 64]
        layer1.add_module('relu1',nn.ReLU(True))
        layer1.add_module('pool1',nn.MaxPool2d(2, 2)) # [100, 32, 32, 32]
        self.layer1 = layer1
        
        layer2 = nn.Sequential()
        layer2.add_module('conv2',nn.Conv2d(32, 64, 3, 1, padding=1)) # [100, 64, 32, 32]
        layer2.add_module('relu2',nn.ReLU(True))
        layer2.add_module('pool2',nn.MaxPool2d(2,2)) # [100, 64, 16, 16]
        self.layer2 = layer2
        
        layer3 = nn.Sequential()
        layer3.add_module('conv3',nn.Conv2d(64, 128, 3, 2, padding=1)) # [100, 128, 8, 8]
        layer3.add_module('relu3',nn.ReLU(True))
        layer3.add_module('pool3',nn.MaxPool2d(2,2)) # [100, 128, 4, 4]
        self.layer3 = layer3
        
        layer4 = nn.Sequential()
        layer4.add_module('fc1',nn.Linear(2048, 512))
        layer4.add_module('fc_relu1',nn.ReLU(True))
        layer4.add_module('fc2',nn.Linear(512,64))
        layer4.add_module('fc_relu2',nn.ReLU(True))
        layer4.add_module('fc3',nn.Linear(64,10))
        self.layer4 = layer4
        
    def forward(self,x):
        self.conv1_output = self.layer1(x)
        self.conv2_output = self.layer2(self.conv1_output)
        self.conv3_output = self.layer3(self.conv2_output)
        self.fc_input = self.conv3_output.view(self.conv3_output.shape[0],-1)
        fc_output = self.layer4(self.fc_input)
        return fc_output
```

```python
data = torch.randn(100,3,64,64)
model = SimpleCNN()
model_out = model(data)

## 验证上面计算的size是否正确
print(model.conv1_output.shape)
print(model.conv2_output.shape)
print(model.conv3_output.shape)
```

### 提取中间层的结构
* 可用children()方法访问当前模型的下一级的所有模块，返回为迭代器。named_children()不仅返回模块的迭代器，还可返回网络名字。
* 用modules()方法会访问当前模型的所有子模块，也即访问到最内层。named_modules()除了返回迭代器，还可返回所有子模块的名字。

```python
list(model.children())
```

```python
net = nn.Sequential(*list(model.children())[:2]) ## 星号可以将列表打散，以单独的元素作为nn.Sequential()的参数赋进来。

## 参考一个例子，如果*号在参数位置，表示接受可变长度的参数，然后将输入的参数以元组的形式组合
def f(*a):
    print(a[0])
    print(a[1])
    print(a[2])

f(*[3,4,5])
```

### 提取参数，自定义初始化
nn.Module()和nn.Sequential()都有两个关于参数的属性，分别是named_parameters()和parameters()。前者给出所有网络层的名字和参数的迭代器，后者给出网络的全部参数的迭代器

```python
for para in model.named_parameters():
    print(para[0])
```

```python
for para in conv_model.named_parameters():
    print(para[0])
```

```python
from torch.nn import init

for m in model.modules():  ## 这里直接用modules，所以直接对m进行isinstance判断
    if isinstance(m,nn.Conv2d):   # m.weight是Variable
        init.normal_(m.weight)    # 这边用m.weight.data也可以，m.weight是nn.Parameter，是默认可autograd的Variable
        init.xavier_normal_(m.weight)   ## xavier_normal初始化非常常用
        init.kaiming_normal_(m.weight)
        m.bias.data.fill_(0)  ## 直接写值进data
    elif isinstance(m,nn.Linear):
        m.weight.data.normal_()
```

```python
## 看一下初始化是否成功
for m in model.modules():
    if isinstance(m,nn.Conv2d):
        print(m.bias.data)
```

```python
## 也可以采用自定义初始化函数，然后将函数apply到model的每一层
def weight_init(m):
    if isinstance(m,nn.Conv2d):
        init.xavier_normal_(m.weight)
        init.constant_(m.bias,1)
    elif isinstance(m,nn.Linear):
        init.normal_(m.weight)

model.apply(weight_init)
```

```python
## 看一下初始化是否成功
for layer in model.named_parameters():
    if 'conv' in layer[0] and 'bias' in layer[0]:
        print(layer[1])
```
