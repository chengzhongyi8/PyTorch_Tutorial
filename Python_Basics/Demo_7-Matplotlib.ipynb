# Matplotlib
### pyplot

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline

x = np.linspace(0,10,1000)
y = np.sin(x)
z = np.cos(x**2)

plt.figure(figsize=(8,4))
plt.plot(x,y,label='$sin(x)$',color='red',linewidth=2)
plt.plot(x,z,'b--',label='$cos(x^2)$')

plt.xlabel('Time(s)')
plt.ylabel('Volt')
plt.title('PyPlot first example')
plt.legend()
```

### subplots

```python
fig = plt.figure(figsize=(8,6))

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

# choose the specific window to plot
ax1.hist(np.random.randn(100),bins=20,color='k',alpha=0.3)
ax2.scatter(np.arange(30),np.arange(30) + 3 * np.random.randn(30))
ax3.plot(np.random.randn(50).cumsum(),'k--')
ax4.plot([1.5,3.5,-2,1.6])
```

```python
# the same as the previous subplotting
fig, ax = plt.subplots(2,2,figsize=(8,6))
ax[0,0].hist(np.random.randn(100),bins=20,color='k',alpha=0.3)
ax[0,1].scatter(np.arange(30),np.arange(30) + 3 * np.random.randn(30))
ax[1,0].plot(np.random.randn(50).cumsum(),'k--')
ax[1,1].plot([1.5,3.5,-2,1.6])
```

### Set ticklabels of axis

```python
plt.plot([1,2,3],[4,5,6])
axis = plt.gca().xaxis
axis.get_ticklocs()
```

```python
axis.get_ticklabels()
```

```python
[x.get_text() for x in axis.get_ticklabels()]  # List comprehension, get the string of the axis
```

```python
plt.plot([1,2,3],[4,5,6])
axis = plt.gca().xaxis
for label in axis.get_ticklabels():
    label.set_color('red')
    label.set_rotation(45)
    label.set_fontsize(12)
    
for line in axis.get_ticklines():
    line.set_color('green')
    line.set_markersize(25)
    line.set_markeredgewidth(3)
```

### Logarithmic plot

```python
plt.figure(figsize=(8,6))
w = np.linspace(0.1, 1000, 1000)
p = np.abs(1/(1+0.1j*w))

plt.subplot(221)
plt.plot(w,p,lw=2)
plt.ylim(0,1.5)  # set the range of y axis

plt.subplot(222)
plt.semilogx(w,p,lw=2)
plt.ylim(0,1.5)

plt.subplot(223)
plt.semilogy(w,p,lw=2)
plt.ylim(0,1.5)

plt.subplot(224)
plt.loglog(w,p,lw=2)
plt.ylim(0,1.5)
```

### Histogram

```python
data = np.loadtxt('china_population.txt')  # [age, populartion of men, population of women]
width = (data[1,0]-data[0,0])*0.4
plt.figure(figsize=(8,5))
plt.bar(data[:,0]-width, data[:,1]/1e7,width,color='b',label='men')
plt.bar(data[:,0], data[:,2]/1e7,width,color='r',label='women')
plt.xlim(-width,100)
plt.xlabel('Age')
plt.ylabel('Population (ten million)')
plt.legend()
```

### Scatter plot

```python
plt.figure(figsize=(8,4))
x = np.random.rand(100)
y = np.random.rand(100)
plt.scatter(x, y, s=x*100, c=y, marker=(5,1), alpha=0.8, lw=2,facecolors='none')
```

### Image visualization and contour

```python
y, x = np.mgrid[-2:2:200j, -2:2:200j]
z = x*np.exp(-x**2-y**2)

extent = [np.min(x), np.max(x), np.min(y), np.max(y)]

plt.figure(figsize=(10,3))
plt.subplot(121)
cs = plt.contour(z,10,extent=extent,colors='k')
plt.clabel(cs,colors='k')
plt.imshow(z,extent=extent,origin='lower')
plt.colorbar()

plt.subplot(122)
plt.imshow(z,extent=extent,cmap=plt.cm.gray,origin='lower')
plt.colorbar()
```

### 3D plotting

```python
import mpl_toolkits.mplot3d

x, y = np.mgrid[-2:2:20j, -2:2:20j]
z = x*np.exp(-x**2-y**2)

plt.figure(figsize=(8,6))
ax = plt.subplot(111,projection='3d')
ax.plot_surface(x,y,z,rstride=2,cstride=1,cmap=plt.cm.Blues)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
